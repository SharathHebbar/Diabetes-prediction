```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('diabetes.csv')
data
```

```python
x = data.drop(['Outcome'],axis=1)
y = data['Outcome']
x,y
```

## Data standardization

```python
scaler = StandardScaler()
scaler.fit(x)
```

#### method 1

```python

xs = scaler.transform(x)
xs
```

#### or

#### method2

```python
x = scaler.fit_transform(x)
x
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y , test_size= 0.2 , stratify = y, random_state=2)
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y , test_size= 0.2)
```

```python
clf = SVC(kernel='linear')
clf.fit(x_train,y_train)

clf1 = SVC(kernel='linear')
clf1.fit(xtrain,ytrain)
```

```python
x_train_pred = clf.predict(x_train)
x_train_acc = accuracy_score(x_train_pred,y_train)
x_train_acc
```

```python
xtrainpred = clf1.predict(xtrain)
xtrainacc = accuracy_score(xtrainpred,ytrain)
xtrainacc
```

```python
x_test_pred = clf.predict(x_test)
x_test_acc = accuracy_score(x_test_pred,y_test)
x_test_acc
```

```python
x_test_pred1 = clf.predict(xtest)
x_test_acc1 = accuracy_score(x_test_pred1,ytest)
x_test_acc1
```
